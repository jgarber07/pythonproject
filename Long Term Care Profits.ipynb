```python
import pandas as pd
import os
import sqlite3
os.getcwd()
```

```python
os.path.isfile('long_term_care_profits.csv')
```

```python
df1 = pd.read_csv('long_term_care_profits.csv')
print(df1)
```

```python
db = sqlite3.connect('lt_care_profits.db')

df1.to_sql('lt_care_profits', db, if_exists='replace')

def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
query = 'select * from lt_care_profits;'
run_query(query)
```

```python
query_table_help = '''
SELECT name , 
*FROM PRAGMA_TABLE_INFO('lt_care_profits');
'''
run_query(query_table_help)
```

```python
query = '''
SELECT * FROM sqlite_master
ORDER BY name;
'''
run_query(query)
```

```python
query_free_profits = '''
SELECT 
inc_tax
,net_income
FROM lt_care_profits
WHERE inc_tax = 0
group by net_income
order by 'net_income' DESC;
'''
run_query(query_free_profits)
```

```python
query_free_average= '''
SELECT 
AVG(net_income) AS AvgTaxFreeIncome 
FROM lt_care_profits
WHERE inc_tax = 0;
'''
run_query(query_free_average)
```

```python
query_profits = '''
SELECT 
inc_tax
,net_income
FROM lt_care_profits
WHERE inc_tax > 0
group by net_income
order by 'net_income' DESC;
'''
run_query(query_profits)
```

```python
query_average= '''
SELECT 
AVG(net_income) AS AvgIncome FROM lt_care_profits
WHERE inc_tax > 0;
'''
run_query(query_average)
```

```python
query_min_max = '''
SELECT MIN(net_income) AS Smallest_Profit
       ,MAX(net_income) AS Highest_Profit
       ,AVG(inc_tax) AS AVGTaxesPaid
FROM lt_care_profits
'''
run_query(query_min_max)
```
